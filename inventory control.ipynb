```python
import pandas as pd
import matplotlib.pyplot as plt
# Read the CSV file into a pandas DataFrame

df = pd.read_csv("New Inventory.csv",)
df
```

```python
print(df.info())
```

```python
df['Inventory Receive'] = pd.to_datetime(df['Inventory Receive'])
```

```python
type(df['Inventory Receive'][0])
```

```python
df['Item Number']=df['Item Number'].str.replace('-','').astype(int)
```

```python
df_cleaned = df.pivot_table(index='Item Number', columns='Inventory Receive',dropna=True,fill_value='0')
df_cleaned.head()
```

```python
# Assuming 'df' is your DataFrame that you want to save

# Save the entire DataFrame to a CSV file
file_path = 'entire_dataframe.csv'  # Replace 'entire_dataframe.csv' with your desired file path and name
df_cleaned.to_csv(file_path, index=True)  # Set index=False to exclude the DataFrame index from the CSV file


```

```python

df_cleaned2 = df.pivot_table(index='Inventory Receive',aggfunc='sum',dropna=True,fill_value='0')
df['Inventory Receive'] = pd.to_datetime(df['Inventory Receive'])
df_cleaned2.head()
```

```python

# Resample the data to get monthly totals
monthly_totals = df_cleaned2.resample('M').sum()

# Plotting the bar chart for monthly totals

plt.figure(figsize=(10, 6))
monthly_totals.plot(kind='bar', figsize=(8,4))
plt.title('Total Monthly Inventory Levels')
plt.xlabel('Month')
plt.ylabel('Total Inventory')
plt.xticks(rotation=45)
plt.legend(title='Inventory Receive')
plt.tight_layout()
plt.show()
```
